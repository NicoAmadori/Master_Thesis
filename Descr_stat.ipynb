```python
import yfinance as yahooFinance
import datetime
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy.stats import norm
import statsmodels.api as sm
from scipy import stats
from scipy.stats import moment
%matplotlib inline
```

```python
start = "2016-01-01"
end = '2023-12-31'
```

```python
XLC = yahooFinance.download('XLC', start=start, end=end)['Adj Close']
XLY = yahooFinance.download('XLY', start=start, end=end)['Adj Close']
XLV = yahooFinance.download('XLV', start=start, end=end)['Adj Close']
XTN = yahooFinance.download('XTN', start=start, end=end)['Adj Close']
XLE = yahooFinance.download('XLE', start=start, end=end)['Adj Close']
XLF = yahooFinance.download('XLF', start=start, end=end)['Adj Close']
FINX = yahooFinance.download('FINX', start=start, end=end)['Adj Close']
```

```python
XLC_returns = XLC.pct_change().dropna()
XLY_returns = XLY.pct_change().dropna()
XLV_returns = XLV.pct_change().dropna()
XTN_returns = XTN.pct_change().dropna()
XLE_returns = XLE.pct_change().dropna()
XLF_returns = XLF.pct_change().dropna()
FINX_returns = FINX.pct_change().dropna()
```

```python
def calculate_statistics(data):
    mean = np.mean(data)
    sd = np.std(data)
    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data)
    data_max = np.max(data)
    data_min = np.min(data)
    ACF = cf_returns = sm.tsa.acf(data, nlags=5)
    ACF_abs = acf_abs_returns = sm.tsa.acf(np.abs(data), nlags=5)
    JB = stats.jarque_bera(data)
    LB_Q = sm.stats.acorr_ljungbox(data,lags=20)
    LB_Q_squared = sm.stats.acorr_ljungbox(np.square(data),lags=20)

    return mean, sd, skewness, kurtosis, data_max, data_min, ACF,ACF_abs, JB, LB_Q, LB_Q_squared
```

```python
XLC_stats = calculate_statistics(XLC_returns)
XLY_stats = calculate_statistics(XLY_returns)
XLV_stats = calculate_statistics(XLV_returns)
XTN_stats = calculate_statistics(XTN_returns)
XLE_stats = calculate_statistics(XLE_returns)
XLF_stats = calculate_statistics(XLF_returns)
FINX_stats = calculate_statistics(FINX_returns)
```

```python
import seaborn as sns
from scipy.stats import norm, t, gennorm
import statsmodels.api as sm
```

```python
start = "2016-01-01"
end = '2023-12-31'
```

```python
XLV = yahooFinance.download('XLV', start=start, end=end)['Adj Close']
XLY = yahooFinance.download('XLY', start=start, end=end)['Adj Close']
XLE = yahooFinance.download('XLE', start=start, end=end)['Adj Close']
```

```python
XLV_returns = XLV.pct_change().dropna()
XLY_returns = XLY.pct_change().dropna()
XLE_returns = XLE.pct_change().dropna()
```

```python
sns.set(style="whitegrid")
plt.figure(figsize=(12, 8))


plt.hist(XLV_returns, bins=500, density=True, alpha=0.6, color='blue', label='Empirical Distribution')

# norm
mu, std = norm.fit(XLV_returns)
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
plt.plot(x, p, 'k', linewidth=2, label='Normal Distribution')

#  t
params_t = t.fit(XLV_returns)
p_t = t.pdf(x, *params_t)
plt.plot(x, p_t, 'r', linewidth=2, label='Student\'s t Distribution')

# skewt
params_skewt = sm.OLS(XLV_returns, sm.add_constant(np.arange(len(XLV_returns)))).fit().params
p_skewt = t.pdf(x, *params_t)
plt.plot(x, p_skewt, 'g', linewidth=2, label='Skewed t Distribution')

# ged
params_ged = gennorm.fit(XLV_returns)
p_ged = gennorm.pdf(x, *params_ged)
plt.plot(x, p_ged, 'purple', linewidth=2, label='Generalized Error Distribution')

plt.title('Histogram of XLV Returns and Model Comparisons')
plt.xlim(-0.2,0.2)
plt.xlabel('Returns')
plt.ylabel('Density')
plt.legend()
plt.show()
```
